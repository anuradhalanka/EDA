### Standardization:

#### task-1:
- step-1: take prevailing wage column : **visa_df['prevailing_wage']**
- step-2: calculate the mean value prevailing wage: **mean=visa_df['prevailing_wage'].mean()**
- step-3: calculate the standard deviation of prevailing wage: **std=visa_df['prevailing_wage'].std()**
- step-4: Calculate the Nr : step-1 - step-2: **Nr= visa_df['prevailing_wage']-mean**
- step-5: divide the step4/step3

```python
mean=visa_df['prevailing_wage'].mean()
std=visa_df['prevailing_wage'].std()
Nr=visa_df['prevailing_wage']-mean
visa_df['prevailing_wage_z']=Nr/std   # prevailing_wage_z is new column
visa_df
```

#### task-2:
- Compare the two columns
- Get the maximum value and minimum value of the indexes
- The both indexes should match

#### idxmax-idxmin:

**returns maximum and minimum value of the indexes**

```python
maxx=visa_df['prevailing_wage'].idxmax(),visa_df['prevailing_wage_z'].idxmax()
minn=visa_df['prevailing_wage'].idxmin(),visa_df['prevailing_wage_z'].idxmin()
maxx,minn
```

**returns maximum and minimum values in those columns**
- Generally Z score values varies from -3 to 3 but you can see some values above 3 and below -3
- Because the values includes outliers in prevailing_wage_z column

```python
max_val=visa_df['prevailing_wage'].max(),visa_df['prevailing_wage_z'].max()
min_val=visa_df['prevailing_wage'].min(),visa_df['prevailing_wage_z'].min()
max_val,min_val
```

#### package name: StandardScaler:
- StandardScaler same as z-score but by using package
- It is under sklearn package
- In the sklearn we have preprocessing
- Read the package
- Save the package
- Apply fit transform
- Compare 3 columns
    - one is original
    - manually we did z-score
    - column with package

```python
from sklearn.preprocessing import StandardScaler   # read the package
ss=StandardScaler()                                # save the package
visa_df['prevailing_wage_ss']=ss.fit_transform(visa_df[['prevailing_wage']])  # apply fit transorm
```

### Normalization:
- Read the data again
- step-1: take the prevailing_wage column: **visa_df['prevailing_wage']**
- step-2: calculate the min value of prevailing_wage: **min=visa_df['prevailing_wage'].min()**
- step-3: caluculate the max value of prevailing_wage: **max=visa_df['prevailing_wage'].max()**
- step-4: calculate the Nr: step-1-step-2
- step-5: DR=step-3-step-2
- step-6: divide the step-4/step-5

```python
x_min=visa_df['prevailing_wage'].min()
x_max=visa_df['prevailing_wage'].max()
Nr=visa_df['prevailing_wage']-x_min
visa_df['prevailing_wage_min_max']=Nr/(x_max-x_min)
visa_df[['prevailing_wage','prevailing_wage_min_max']]
```

#### package name: MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
visa_df['prevailing_wage_min_max_ss']=mms.fit_transform(visa_df[['prevailing_wage']])
visa_df[['prevailing_wage','prevailing_wage_min_max','prevailing_wage_min_max_ss']]
```
