#### take :

- If you want to get middle data
- Take option is used to extract the particular index of data
- take has one parameter called axis
- axis=1 represents columns
- axis=0 represents rows
- bydefault axis=0 is there
- in below example we give 100,200,300 and we did not provide axis values
- I understood we are talking about rows as axis=0 bydefault

```python
visa_df.take([100,200,300])   # returns row number 100,200,300
```

```python
visa_df.take([100,200,300],axis=1)  # gives error as axis=1 means columns and we don't have 100,200,300 column numbers
# bydefault axis=0  means rows, if you do shift+tab  in () you can see that
```

```python
visa_df.take([2,5,7],axis=1) # gives o/p as column numbers 3rd,6th,8th as python index starts with 0
```

```python
visa_df.take[100:105]   # slicing won't work here, gives error 
```

$Note$ :
- we can't take rows and columns at a time in one single take

```python
# Question:
# I want to take rows: 100,200,300 and columns: 2,5,7
# with single syntax not possible
# need to use 2 takes, one for rows and second for columns 
# and need to mention axis=1 for columns

visa_df.take([100,200,300]).take([2,5,7],axis=1)
```

#### iloc :
- to solve the problem in take we use iloc
- rows and columns at a time is possible using iloc

```python
# syntax:
# visa_df.iloc[<rows>,<columns>]
# visa_df.iloc[<start:end>,<start:end>]
# rows=[100,200,300]
# columns=[2,5,7]
```

```python
visa_df.iloc[[100,200,300],[2,5,7]]
```

```python
visa_df.iloc[[100,200,300]]  # same like take it works
```

```python
visa_df.iloc[100:105,[2,5,7]]   # slicing also working
```

```python
visa_df.iloc[100:105,2:5]
# returns 100 to 104 row numbers and 3rd,4th,5th columns as python index starts from 0
```

```python
visa_df.iloc[100:105,2]
```
