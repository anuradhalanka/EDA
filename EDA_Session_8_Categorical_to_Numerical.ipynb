### map method :
- First get the unique labels for each column
- Create a dictionary by providing a values to each label
- For example case_status column has two labels ('Certified','Denied'), we are providing two values (0,1)
    - Assign 0 for Certified
    - Assign 1 for Denied

```python
visa_df['case_status'].unique()    # to get unique values in case_status column
```

```python
d={'Denied':1,'Certified':0}
d      # creating dictionary by providing values to each label
```

```python
visa_df['case_status_new']=visa_df['case_status'].map(d)       # using map method converting categorical to numerical values
visa_df
# case_status_new is new column created, with labels replaced with dictionary values
# old column case_status also will be there
```

**task**
- Read the data again
- Apply for loop
- Iterate through columns
- Get the unique labels of each column
- Get the count of unique labels
- Create a dictionary
- Apply in to the map

```python
# to convert all the categorical columns into numerical columns in the visa_df DataFrame

visa_df=pd.read_csv(r"C:\Users\Adireddy\Documents\Naresh IT Data Science & AI\Data files\visadataset - visadataset.csv")
cols=visa_df.select_dtypes(include='object').columns    # to get categorical columns
for i in cols:
    labels=list(visa_df[i].unique())        # unique gives o/p in the form of array, you need to convert into list by using list()
    values=[i for i in range(len(labels))]  # using list comprehension iterate through each unique label
    d=dict(zip(labels,values))              # to create labels, values pairs using dictionary
    visa_df[i]=visa_df[i].map(d)            # using map method all the categorical columns convert to numerical
visa_df
```

### np.where method :
- np.where applicable only for binary
- binary means columns having only two unique values like Y/N, Certified/Denied
- np.where will take 3 arguments
    - condition
    - True value
    - False value

**scenario:**
- For example case_status value equal to 'Certified' replace with zero
- otherwise replace with 1

```python
visa_df=pd.read_csv(r"C:\Users\Adireddy\Documents\Naresh IT Data Science & AI\Data files\visadataset - visadataset.csv")
con=visa_df['case_status']=='Certified'    # condition
visa_df['case_status']=np.where(con,0,1)   # np.where(condition,True value,False value)
visa_df
```

#### One hot encoder:
- One hot means if one is ON another one is OFF
- ON represents with 1, and OFF represents with 0
- For example case_status column has two unique lables
    - Denied
    - Certified
- If we apply One hot encoder on case_status column it will create two extra columns
    - case_status_Denied
    - case_status_Certified
- At a time one value in a column is 1, remaining all values in that column are 0

| case_status | case_status_Denied | case_status_Certified |
|---------|----|----|
|Denied|1|0|
|Certified|0|1|

####  Advantages of One hot encoder:
- In machine learning we have one important property is **independent of variables**
- Machine learning models expects variables are **independent each other**
- One hot encoder creating new columns, the new columns are **independent each other**
- Independent variables means both variables have **90 degrees phase shift** - 90 degrees phase shift also called as **perpendicular each other** - perpendicular means **orthogonal each other**

#### Disadvantages:
- One hot encoder creates new columns equal to number of unique lables
- If number of unique lables increase, new columns also increase
- The more columns leads data is more
- Then data is more the complexity will increase
- complexity increases more computation power required
- Also time increase
- The entire process names as **curse of dimensionality**
