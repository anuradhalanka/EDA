#### Data Transformations:
- Generally data has 3 types of distributions
    - Normal Distribution
    - Positive Skewed
    - Negative Skewed
- All the maths concepts developed by assumption as data follows normal distribution
- Data skew happens because of many reasons, in that one reason is outliers
- Once we perform the outlier analysis then we can apply transformation methods, if the data does not follows normality
- We have some transformation methods mentioned below:
    - Log transformation
    - Exponential transformation
    - Reciprocal transformation
    - Square root transformation
    - Power transformation

### Exponential Transformation:

#### Step-1:
#### import required packages:

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Step-2
#### Read the data:

```python
exp_data=np.random.exponential(size=10000)
exp_data
```

#### Step-3:
#### plot the data:

```python
plt.hist(exp_data,bins=20,label='Exponential')   # histogram of exp_data, bins=intervals
plt.legend()  # Exponential will disply at the corner of histogram
plt.show()
```

#### Normal Data:

```python
norm_data=np.random.normal(size=10000)
norm_data
```

#### plot the normal data:

```python
plt.hist(norm_data,bins=20,label='Normal')
plt.legend()
plt.show()
```

#### Goal:
#### Convert exp_data to normal distributuion:
- bins means intervals

#### Step-4:
#### Log Transformation:
- log transformation means performing logarithm operations on original data
- It is an approach to convert into normal distribution
- log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)    # performing logarithm operation on exp_data 
log_data
```

```python
# to plot logarithm data
plt.hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.show()
```

#### comparing exponential and logarithm plots:

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Logarithm')
plt.legend()
plt.show()

# converted exp_data to normal distribution
```

#### Reciprocal transformation:
- for x, reciprocal is 1/x
- reciprocal will fail when values are low
- for example x=0 then 1/0 will fail
- Reciprocal does not give normal distribution

```python
x=2
1/x
```

```python
resiprocal_data=np.reciprocal(exp_data)
plt.hist(resiprocal_data,bins=20,label='Reciprocal')
plt.legend()
plt.show()
```

#### Square Root Transformation:

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

#### comparing exponential, logarithm, reciprocal, square root transformations plots:

```python
plt.figure(figsize=(12,3))
exp_data=np.random.exponential(size=10000)
plt.subplot(1,4,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
log_data=np.log(exp_data)  # to convert exp_data to normal distribution
plt.subplot(1,4,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
resiprocal_data=np.reciprocal(exp_data)
plt.subplot(1,4,3).hist(resiprocal_data,bins=20,label='Reciprocal')
plt.legend()
square_root_data=np.sqrt(exp_data)
plt.subplot(1,4,4).hist(square_root_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

#### Power Transformation:
- box-cox method
- yeo-johnson method
- It is under sklearn preprocessing
- Yeo-Johnson can handle datasets with both positive and negative values
- box-cox method will apply only for positive data
- with box_cox method we'll get perfect normal distribution plot out of all

#### By applying ye-johnson method on PowerTransformer package:

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer   # read the package
power_yeo=PowerTransformer(method='yeo-johnson')         # save the package
exp_data1=exp_data.reshape(-1,1)                     # to get each value in one list
data_trans_yeo=power_yeo.fit_transform(exp_data1)             # apply fit transform
plt.hist(data_trans_yeo,bins=20,label='Yeo-Johnson')                                 # to plot the transformed data
plt.legend()
plt.show()
```

#### Applying Box Cox method on Power Transformer package:

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer    # read the package
power_boxcox=PowerTransformer(method='box-cox') # save the package
exp_data1=exp_data.reshape(-1,1)
data_trans_boxcox=power_boxcox.fit_transform(exp_data1)           # apply fit transform
plt.hist(data_trans_boxcox,bins=20,label='Box-Cox')                                 # to plot the transformed data
plt.legend()
plt.show()
```

#### Comparing yeo-johnson and box-cox in power transformations:

```python
from sklearn.preprocessing import PowerTransformer
power_yeo=PowerTransformer(method='yeo-johnson')
data_trans_yeo=power_yeo.fit_transform(exp_data.reshape(-1,1))

power_boxcox=PowerTransformer(method='box-cox')
data_trans_boxcox=power_boxcox.fit_transform(exp_data.reshape(-1,1))

plt.figure(figsize=(20,5))
plt.subplot(1,2,1).hist(data_trans_yeo,bins=20,label='Yeo-Johnson')
plt.legend()
plt.subplot(1,2,2).hist(data_trans_boxcox,bins=20,label='Box-Cox')
plt.legend()
plt.show()
```
