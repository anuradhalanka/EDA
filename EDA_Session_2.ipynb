#### create dataframe using dictionary :

```python
name=['Avinash','Aakash','Aadhya']
age=[25,30,35]
pd.DataFrame(zip(name,age),columns=['Names','Ages'])
```

```python
dict1={'Names':['Avinash','Aakash','Aadhya'],'Ages':[25,30,35]}  # key,value pair
dict1
```

```python
pd.DataFrame(dict1)
# keys becomes columns
# values becomes rows
# no need to provide column names seperately
```

```python
dict2={'Names':'Avinash',
       'Ages':25,
       'City':'Hyd'}
pd.DataFrame(dict2)
# dict1 has list
# dict2 don't has list
# value error will come if you don't pass index
```

```python
dict2={'Names':'Avinash',
       'Ages':25,
       'City':'Hyd'}
pd.DataFrame(dict2,index=[1,2,3])
```
